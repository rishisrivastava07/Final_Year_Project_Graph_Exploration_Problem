   - In this approach, each node gathers information from its immediate neighbors using only local knowledge.

   - Here, each node can see the status and information of its immediate neighbors.
   - The decision for the node color is made based on one-hop visibility.

# **Rooted Configuration for 0-hop Visibility**

### High-level overview:

---


On a high level, MIS-GraphRooted follows
a DFS-like exploration of the graph, while finding back-edges of the DFS by using the unsettled agents as scouts. The scouts visit the neighbors and report back the status of the neighbors. We start with the root being part of the MIS (the settled agent takes color ùëüùëíùëë). On visiting a new node, we determine whether that node is part of MIS or not, by scouting all its neighbors to check if that contains an agent that is part of MIS. If no neighbor is part of MIS, then the node becomes a part of MIS. Otherwise, it does not (becomes color ùëèùëôùë¢ùëí). In Figure 1, we show the stages of algorithm MIS-GraphRooted.

image.png

G = (ùëâ, ùê∏), where |ùëâ |= ùëõ and |ùê∏|= ùëö.

Nodes are indistinguishable from each other.

The edges at a node ùë¢ are uniquely labeled with ports
{0, . . . , ùõøùë¢‚àí1},

where ùõøùë¢ is the degree of node ùë¢.

Œî is the maximum degree of the graph.

Initially, ùëõ agents are located in the graph arbitrarily.

## **For K = 0 (Zero Hop Visibility) :-**
---
### ***Time Complexity :*** O(n‚àÜ)
### ***Space Complexity :*** Œò(log n)

## ***Approach - 01 (Wihtout Saving)***

````python
from collections import deque
import networkx as nx
import matplotlib.pyplot as plt

class Node:
    def __init__(self, name):
        self.name = name
        self.state = None  # Scout or not
        self.phase = "forward"  # "forward" or "backtrack"
        self.settled = False
        self.color = "black"  # Red (MIS), Blue (Non-MIS), Black (Unvisited)
        self.scoutport = None  # Port to scout next
        self.scoutparent = None  # Port at neighbor leading to this node
        self.target = None  # Information about the scouted node
        self.recent = 0  # Tracks last explored port
        self.next = None  # Next smallest available port to unvisited node

def initialize_graph():
    adjacency_list = {
        "t": {"x": 0, "w": 1},
        "u": {"v": 0, "w": 1, "y": 2},
        "v": {"u": 0, "w": 1, "x": 2, "z": 3},
        "w": {"t": 0, "u": 1, "v": 2, "x": 3, "z": 4},
        "x": {"t": 0, "v": 1, "w": 2, "y": 3},
        "y": {"u": 0, "x": 1, "z": 2},
        "z": {"v": 0, "w": 1, "y": 2},
    }
    return {node: Node(node) for node in adjacency_list}, adjacency_list

def draw_graph(nodes, adjacency_list, step, processing_edge=None, fixed_pos=None):
    G = nx.Graph()
    edge_colors = {}

    for node, neighbors in adjacency_list.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)
            edge_colors[(node, neighbor)] = "black"  # Default color

    if processing_edge:
        edge_colors[processing_edge] = "red"  # Highlight current processing edge

    color_map = {"red": "red", "blue": "blue", "black": "gray"}
    node_colors = [color_map[nodes[n].color] for n in G.nodes()]
    edge_color_list = [edge_colors.get((u, v), edge_colors.get((v, u), "black")) for u, v in G.edges()]

    plt.figure(figsize=(6, 6))
    if fixed_pos is None:
        fixed_pos = nx.spring_layout(G, seed=42)  # Fixed layout for consistency
    nx.draw(G, pos=fixed_pos, with_labels=True, node_color=node_colors, edge_color=edge_color_list, node_size=1500, font_size=10, edge_cmap=plt.cm.Reds)
    plt.title(f"Graph Representation at Step {step}")
    plt.show()
    return fixed_pos

def print_table(nodes, step):
    print(f"\nStep {step}: Node State Table")
    print("-" * 100)
    print(f"{'Node':<5} {'State':<10} {'Phase':<10} {'Color':<10} {'ScoutPort':<10} {'ScoutParent':<12} {'Target':<10} {'Recent':<7} {'Next':<5}")
    print("-" * 100)
    for node in nodes.values():
        print(f"{node.name:<5} {str(node.state):<10} {node.phase:<10} {node.color:<10} {str(node.scoutport):<10} {str(node.scoutparent):<12} {str(node.target):<10} {str(node.recent):<7} {str(node.next):<5}")
    print("-" * 100 + "\n")

def find_mis(nodes, adjacency_list, root="t"):
    fixed_pos = draw_graph(nodes, adjacency_list, step=0)  # Initial graph configuration
    stack = [(root, None)]  # Using stack for DFS-like traversal
    nodes[root].settled = True
    nodes[root].color = "red"  # Root is always part of MIS
    print(f"Starting at root node '{root}', marking it as RED (MIS).\n")

    step = 0  # Step counter
    while stack:
        step += 1
        current, parent = stack.pop()
        node = nodes[current]
        print(f"Step {step}: Processing node '{current}' (Parent: {parent if parent else 'None'})")

        # Set node attributes
        node.state = "scout"
        node.scoutport = 0 if parent is None else adjacency_list[parent][current]
        node.scoutparent = None if parent is None else adjacency_list[current][parent]

        print(f" - Scouting neighbors of '{current}'...")
        reports = []  # Collect reports from neighbors
        unvisited_neighbors = []

        # Sort neighbors based on smallest port number
        sorted_neighbors = sorted(adjacency_list[current], key=lambda x: adjacency_list[current][x])

        for neighbor in sorted_neighbors:
            if neighbor == parent:
                continue  # Do not skip the parent

            if not nodes[neighbor].settled:
                nodes[neighbor].parent = current
                unvisited_neighbors.append(neighbor)
                print(f"   -> Found unvisited neighbor: {neighbor}")
            else:
                reports.append(nodes[neighbor].color)
                print(f"   -> Neighbor {neighbor} already settled, reports {nodes[neighbor].color}")

        if not unvisited_neighbors and all(nodes[n].settled for n in nodes):
            print(f" - No unvisited nodes left, terminating.")
            break

        for neighbor in reversed(unvisited_neighbors):
            stack.append((neighbor, current))

        has_red_neighbor = any(nodes[neighbor].color == "red" for neighbor in adjacency_list[current] if nodes[neighbor].settled)

        if has_red_neighbor:
            node.color = "blue"
            print(f" - Node '{current}' has a RED neighbor, marking it BLUE.")
        else:
            node.color = "red"
            print(f" - Node '{current}' has NO RED neighbor, marking it RED.")

        node.state = "settled"
        node.phase = "backtrack" if not unvisited_neighbors else "forward"
        node.target = "red" if node.color == "red" else "blue"
        node.recent += len(unvisited_neighbors)
        node.settled = True

        print_table(nodes, step)
        draw_graph(nodes, adjacency_list, step, processing_edge=(parent, current) if parent else None, fixed_pos=fixed_pos)

    mis_set = {node.name for node in nodes.values() if node.color == "red"}
    print("Final Maximum Independent Set (MIS):", mis_set)
    return mis_set

nodes, adjacency_list = initialize_graph()
mis_set = find_mis(nodes, adjacency_list)
````

## ***Approach - 02 (Saving as Output)***

````python
import os
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.table import Table

class Node:
    def __init__(self, name):
        self.name = name
        self.settled = False
        self.color = "black"  # Red (MIS), Blue (Non-MIS), Black (Unvisited)
        self.state = None
        self.phase = "forward"
        self.scoutport = None
        self.scoutparent = None
        self.target = None
        self.recent = 0
        self.next = None

def initialize_graph():
    adjacency_list = {
        "t": {"x": 0, "w": 1},
        "u": {"v": 0, "w": 1, "y": 2},
        "v": {"u": 0, "w": 1, "x": 2, "z": 3},
        "w": {"t": 0, "u": 1, "v": 2, "x": 3, "z": 4},
        "x": {"t": 0, "v": 1, "w": 2, "y": 3},
        "y": {"u": 0, "x": 1, "z": 2},
        "z": {"v": 0, "w": 1, "y": 2},
    }
    return {node: Node(node) for node in adjacency_list}, adjacency_list

def save_graph(nodes, adjacency_list, step, current_node=None, description=""):
    folder = "MIS_Rooted"
    os.makedirs(folder, exist_ok=True)  # Ensure directory exists

    G = nx.Graph()
    edge_colors = {}

    for node, neighbors in adjacency_list.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)
            edge = tuple(sorted([node, neighbor]))
            edge_colors[edge] = "gray"

    if current_node:
        for neighbor in adjacency_list[current_node]:
            edge = tuple(sorted([current_node, neighbor]))
            edge_colors[edge] = "red"

    color_map = {"red": "red", "blue": "blue", "black": "gray"}
    node_colors = [color_map[nodes[n].color] for n in G.nodes()]

    fig, ax = plt.subplots(figsize=(12, 8))
    pos = nx.spring_layout(G, seed=42)

    nx.draw(G, pos=pos, with_labels=True, node_color=node_colors, node_size=1500, font_size=10,
            edge_color=[edge_colors[tuple(sorted(edge))] for edge in G.edges()])

    plt.title(f"Step {step}: {description}", fontsize=14, fontweight='bold', pad=20)

    main_table_data = [["Node", "State", "Phase", "Color", "ScoutPort", "ScoutParent", "Target", "Recent", "Next"]]
    for node in nodes.values():
        main_table_data.append([
            node.name, str(node.state), node.phase, node.color, str(node.scoutport),
            str(node.scoutparent), str(node.target), str(node.recent), str(node.next)
        ])

    main_table = Table(ax, bbox=[-0.6, -0.3, 0.9, 0.25])

    for i, row in enumerate(main_table_data):
        for j, cell in enumerate(row):
            cell_obj = main_table.add_cell(i, j, width=0.11, height=0.05, text=cell, loc='center',
                                           facecolor='#d9edf7' if i == 0 else 'white', edgecolor='black')
            cell_obj.get_text().set_fontsize(10)
    ax.add_table(main_table)

    if current_node:
        local_info = [["Current Node", current_node]]
        local_info.append(["Parent", str(nodes[current_node].scoutparent)])
        for neighbor in adjacency_list[current_node]:
            target_color = nodes[neighbor].color if nodes[neighbor].settled else "black (Empty)"
            local_info.append([f"{neighbor}.target", target_color])
        local_info.append(["Final Decision", f"{current_node}.color = {nodes[current_node].color}"])

        local_table = Table(ax, bbox=[0.5, -0.3, 0.4, 0.25])
        for i, row in enumerate(local_info):
            for j, cell in enumerate(row):
                cell_obj = local_table.add_cell(i, j, width=0.2, height=0.05, text=cell, loc='center',
                                                facecolor='#d9edf7' if i == 0 else 'white', edgecolor='black')
                cell_obj.get_text().set_fontsize(10)
        ax.add_table(local_table)

    plt.savefig(os.path.join(folder, f"step_{step}.png"), bbox_inches="tight")
    plt.close()

def find_mis(nodes, adjacency_list, root="t"):
    step = 0
    save_graph(nodes, adjacency_list, step, description="Initial Step: All nodes are unvisited.")

    stack = [(root, None)]

    nodes[root].settled = True
    nodes[root].color = "red"
    nodes[root].state = "settled"

    while stack:
        step += 1
        current, parent = stack.pop()
        node = nodes[current]

        if not node.settled:
            node.state = "scout"

        node.scoutport = 0 if parent is None else adjacency_list[parent][current]
        node.scoutparent = parent

        unvisited_neighbors = sorted(
            [n for n in adjacency_list[current] if not nodes[n].settled],
            key=lambda x: adjacency_list[current][x]
        )

        for neighbor in adjacency_list[current]:
            if nodes[neighbor].settled:
                nodes[current].target = nodes[neighbor].color

        if node.color == "black":
            has_red_neighbor = any(nodes[neighbor].color == "red" for neighbor in adjacency_list[current] if nodes[neighbor].settled)
            node.color = "blue" if has_red_neighbor else "red"
            node.settled = True
            node.state = "settled"

        node.phase = "backtrack" if not unvisited_neighbors else "forward"
        node.recent += len(unvisited_neighbors)

        save_graph(nodes, adjacency_list, step, current_node=current, description=f"Processing node '{current}'")

        for neighbor in reversed(unvisited_neighbors):
            stack.append((neighbor, current))

    save_graph(nodes, adjacency_list, step + 1, description="Final Step: MIS computation completed.")

    mis_set = {node.name for node in nodes.values() if node.color == "red"}
    print("Final Maximum Independent Set (MIS):", mis_set)
    return mis_set

# Run the algorithm
nodes, adjacency_list = initialize_graph()
mis_set = find_mis(nodes, adjacency_list)
````

# **Dispered Configuration for 1-hop Visibility**

### High-level overview:
---
Each agent executing MIS-GraphDispersed performs a neighborhood visit to determine if it is the largest ID agent in the neighborhood. The agents follow two rules to determine whether they are part of MIS or not,

1.  if an agent has a neighbor that is part of the MIS (has color ùëüùëíùëë), then it does not become part of MIS (takes color ùëèùëôùë¢ùëí), and

2.  if an agent determines that all larger ID agents in its neighborhood are not part of the MIS, then it becomes part of the MIS.

As a consequence of these rules, the largest ID agent among all agents becomes part of MIS (sets color ùëüùëíùëë). Subsequently, other agents decide whether they are part of the MIS or not based on their neighbors. In one phase, an agent can visit one neighbor and find it in at most 2‚Ñì rounds. To visit all its neighbors, it needs at most 2Œî‚Ñì rounds. The algorithm has mega-phases of size 2Œî‚Ñì, that each agent can keep track of by having the knowledge of ùëõ and Œî, since ‚Ñì = ùëê ‚åàlogùëõ‚åâ for some constant ùëê. At the end of each mega-phase, an agent can determine its color if one of the conditions is satisfied, otherwise it continues to the next mega-phase. Once an agent makes a decision about its color, it remains stationary thereafter. In Figure, we show the assignment of colors for each agent in successive mega-phases.

## **For K = 1 (One Hop Visibility) : -**
---
### ***Time Complexity :*** O(nŒîlog n)
### ***Space Complexity :*** Œò(log n)

## ***Approach - 01 (Wihtout Saving)***

````python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Initialize graph
graph = {
    't': {'x', 'w'},
    'u': {'v', 'w', 'y'},
    'v': {'u', 'w', 'x', 'z'},
    'w': {'t', 'u', 'v', 'x', 'z'},
    'x': {'t', 'v', 'w', 'y'},
    'y': {'u', 'x', 'z'},
    'z': {'v', 'w', 'y'}
}

# Function to plot graph and display table below the image
def plot_graph_and_table(graph, mis_set, removed_nodes, step, pos, df):
    fig, ax = plt.subplots(figsize=(8, 8))
    G = nx.Graph()
    for node, neighbors in graph.items():
        for neighbor in neighbors:
            G.add_edge(node, neighbor)

    colors = []
    for node in G.nodes():
        if node in mis_set:
            colors.append('red')  # Selected in MIS
        elif node in removed_nodes:
            colors.append('blue')  # Removed
        else:
            colors.append('gray')  # Default

    nx.draw(G, pos, with_labels=True, node_color=colors, edge_color='black', node_size=500, ax=ax)
    plt.title(f"Step {step}")

    # Display table below the graph with borders
    df.fillna('NULL', inplace=True)
    table_text = df.to_markdown(index=False, tablefmt='grid')
    fig.text(0.5, -0.4, table_text, fontsize=12, family='monospace', verticalalignment='bottom', horizontalalignment='center')

    plt.subplots_adjust(bottom=0.4)  # Add more space between graph and table
    plt.show()

    # Save table as CSV
    df.to_csv(f'mis_step_{step}.csv', index=False)

# Step 2: Assign increasing IDs to nodes
node_ids = {node: i for i, node in enumerate(sorted(graph.keys()))}

# Step 3: Sort nodes by highest ID
sorted_nodes = sorted(node_ids.keys(), key=lambda node: node_ids[node], reverse=True)

# Step 4: Define graph layout for consistency
G_layout = nx.Graph()
for node, neighbors in graph.items():
    for neighbor in neighbors:
        G_layout.add_edge(node, neighbor)
pos = nx.spring_layout(G_layout, seed=42)

# Step 5: Initialize tracking variables
mis_set = set()
removed_nodes = set()
step = 1
node_states = {node: 'black' for node in graph}  # Initially all black
parent = {node: '-' for node in graph}  # No parent initially

# Display initial configuration table
data = []
for node in graph:
    highest_candidate = max(graph[node], key=lambda n: node_ids[n], default='NULL')
    available_neighbors = ', '.join(graph[node]) if graph[node] else 'NULL'
    data.append([node, 'black', highest_candidate, available_neighbors, 'NULL'])
df = pd.DataFrame(data, columns=["Node", "State", "Highest Candidate", "Available Neighbors", "Parent"])
plot_graph_and_table(graph, mis_set, removed_nodes, "Initial Configuration", pos, df)

# Step 6: Process nodes
while sorted_nodes:
    highest_node = sorted_nodes.pop(0)  # Pick the highest ID node
    if highest_node in removed_nodes:
        continue

    mis_set.add(highest_node)
    removed_nodes.add(highest_node)
    node_states[highest_node] = 'red'  # Mark as MIS

    # Get available neighbors
    available_neighbors = [neighbor for neighbor in graph[highest_node] if neighbor not in removed_nodes]

    # Update parent
    for neighbor in available_neighbors:
        if parent[neighbor] == '-':
            parent[neighbor] = highest_node

    # Mark neighbors as blue and remove them from consideration
    for neighbor in graph[highest_node]:
        if neighbor not in removed_nodes:
            removed_nodes.add(neighbor)
            node_states[neighbor] = 'blue'

    # Remove blue nodes from sorted_nodes
    sorted_nodes = [node for node in sorted_nodes if node not in removed_nodes]

    # Create and display table
    data = []
    for node in graph:
        highest_candidate = max(graph[node], key=lambda n: node_ids[n], default='NULL')
        available_neighbors = ', '.join([n for n in graph[node] if n not in removed_nodes]) if any(n not in removed_nodes for n in graph[node]) else 'NULL'
        data.append([node, node_states[node], highest_candidate, available_neighbors, parent[node] if parent[node] != '-' else 'NULL'])

    df = pd.DataFrame(data, columns=["Node", "State", "Highest Candidate", "Available Neighbors", "Parent"])

    # Plot graph and display table below the image
    plot_graph_and_table(graph, mis_set, removed_nodes, step, pos, df)
    step += 1
````
